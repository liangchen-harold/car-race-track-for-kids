```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import crtk

records = []
records += crtk.analyze('data/20230806-1150-0.csv.gz')
records += crtk.analyze('data/20230806-1150-1.csv.gz')

df = pd.DataFrame(records, columns=['experiment', 'speed'])
# display(df)

sns.scatterplot(data=df, y='speed', x='experiment')
plt.show()

sns.boxplot(data=df, y='speed', x='experiment')
sns.stripplot(data=df, y='speed', x='experiment', color = 'black', alpha = 0.3)
plt.show()
```
