```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import crtk

# read data from log

df = pd.read_csv('data/20230806-0933.csv.gz', header=None)

# plot all data

# crtk.show_ts(df)
crtk.show(df)


# analyze every detection

tri = crtk.Trigger()

for i, (a, b, ts) in df.iterrows():
    # print(a, b, ts)
    dur = tri.check(a, b, ts)
    if dur is not None:
        print("dur = {} ms, speed = {:.4f} km/h".format(dur/1000, 0.2 / (dur/1000_000) * 3.6))
        crtk.show(df[i-1000:i+500])
```
